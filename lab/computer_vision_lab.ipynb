````python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import mnist
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns

# Q1
df = mnist.load_data('mnist.db')
train, test = df
X_train, y_train = train
X_test, y_test = test

np.set_printoptions(edgeitems=30, linewidth=100000)
for i in range(5): 
    print(y_test[i], '\n') 
    print(X_test[i], '\n')
    plt.contourf(np.rot90(X_test[i].transpose()))  # Make a contour plot of the matrix values
    plt.show()
````

````python
# Q2
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
````

````python
#Q3
X_train = X_train.reshape(X_train.shape[0], 28*28)
X_test = X_test.reshape(X_test.shape[0], 28*28)
````

````python
# Q4
k_values = [1, 3, 5, 7, 9]
best_k = None
best_accuracy = 0

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy for k={k}: {accuracy:.4f}")
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_k = k

print(f"Best k: {best_k} with accuracy: {best_accuracy:.4f}")
````

````python
# Q5 & 6
knn_final = KNeighborsClassifier(n_neighbors=best_k)
knn_final.fit(X_train, y_train)
y_pred_final = knn_final.predict(X_test)

conf_matrix = confusion_matrix(y_test, y_pred_final)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

print("Classification Report:")
print(classification_report(y_test, y_pred_final))
````

The optimal number of neighbors was found to be 3 with a accuracy of 97.5%, this value minimized errors while maintaining computational efficiency. The final accuracy score on the test set was measured, and it reflected how well the model was able to correctly classify handwritten digits. The higher the accuracy, the better the classifier performed in recognizing handwritten digits correctly. The model’s accuracy depends on the quality of feature extraction and the chosen k value, as too few or too many neighbors may lead to overfitting or underfitting.

A common mistakes from the matrix are, digits that look visually similar (4 and 9) are likely to be confused with each other. This happens because some handwritten styles make digits appear ambiguous. Interesting patterns observed were that some digits are almost always classified correctly (0). Certain numbers with more variation in how people write them tend to be more frequently misclassified (5 and 3). Misclassification is more common in digits with overlapping pixel patterns.
